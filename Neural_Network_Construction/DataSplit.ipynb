```python
import json
import random

with open('FalsePositives.json', 'r') as f: 
    fp_data = json.load(f);
    

img_names = []

train_data = {}
val_data = {}
test_data = {}

x = True;

for key in fp_data.keys():
    img_names.append(key)

    
train_split = int(0.70 * len(fp_data)) + 1
val_split = train_split + int(0.15 * len(fp_data)) 

for i in range(0, train_split): 
    train_data[img_names[i]] = fp_data[img_names[i]]
    
for i in range(train_split, val_split):
    val_data[img_names[i]] = fp_data[img_names[i]]
    
for i in range(val_split, len(fp_data)):
    test_data[img_names[i]] = fp_data[img_names[i]]

write_data = json.dumps(train_data)
    
with open('FPTrainData.json', 'w') as f:
    f.write(write_data)
    
write_data = json.dumps(val_data)

with open('FPValData.json', 'w') as f:
    f.write(write_data)
    
write_data = json.dumps(test_data)

with open('FPTestData.json', 'w') as f:
    f.write(write_data)
    
```
